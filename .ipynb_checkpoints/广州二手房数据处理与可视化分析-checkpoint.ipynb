### 2)查看统计重复数据

```python
df_all.duplicated().sum()
```

```python
df_all.户型
```

```python
#（1）提取室
df_all['室']=df_all['户型'].str.split('|').str[0].str.extract(r'(\d)室')
df_all.室
```

```python
#（2）提取厅
df_all['厅']=df_all['户型'].str.split('|').str[0].str.extract(r'(\d)厅')
df_all.厅
```

```python
#(3)删除houseInfo列
df_all=df_all.drop('户型',axis=1)
df_all
```

```python
df_all.面积
```

```python
#（4）提取面积
df_all['面积']=df_all['面积'].str.split('|').str[0].str.extract(r'(\d+.*\d+)平米')
df_all.面积
```

```python
df_all.楼层
```

```python
#（5）将楼层列处理为：高层'、低层、中层、地下室'
def transform_floor(x):
    if x=='高楼层' or x=='顶层' or x=='上叠':
        return '高层'
    elif x=='低楼层' or x=='低层' or x=='下叠' or x=='1层' or x=='2层' or x=='3层':
        return '低层'
    elif x=='中楼层' or x=='4层' or x=='5层' or x=='6层':  
         return '中层'
    elif x=='地下室' : 
        return '地下室'
    else:
        return '高层'
```

```python
#（6）将几个数据一般化（楼层、朝向、年份、板塔）
df_all['楼层信息']=df_all['楼层'].str.replace(r'\(.*?\)','').str.strip()
df_all['楼层信息']=df_all.楼层.apply(transform_floor)
df_all=df_all.drop('楼层',axis=1)
df_all.head()
```

```python
#（7）转换数据类型（总价和单价转换为float）其他转换为整型
df_all['室']=df_all['室'].astype('int')
df_all['厅']=df_all['厅'].astype('int')
df_all['面积']=df_all['面积'].astype('float')
df_all.info()
```

```python
#(8)查看是否有空值
df_all.isnull().sum()
```

```python
#(9)空值直接删除
df_all=df_all.dropna()
df_all
```

### 3）图表分析

```python
#(10)按照总价绘制饼图
#设置分割点
bins=[50,200,400,800,1000,4500]

#对总价进行分割（分箱)
bins_label=['200万及以下','200-400万','400-800万','800-1000万','1000万以上']
df_all['pricecut']=pd.cut(df_all['总价格（万）'],bins,right=False,labels=bins_label)

#对分箱后的数据进行，分组、统计、排序
df_all['pricecut'].value_counts()

#转换成列表对值
price_num=df_all['pricecut'].value_counts()
data_pair=[list(z) for z in zip(price_num.index.tolist(),price_num.values.tolist())]
#data_pair

#绘制饼图
pie1= Pie(init_opts=opts.InitOpts(width='800px', height='750px'))
pie1.add('', data_pair=data_pair, radius=['30%', '60%'], rosetype='radius')
pie1.set_global_opts(title_opts=opts.TitleOpts(title='广州二手房都处在哪些价位？'), 
                     legend_opts=opts.LegendOpts(orient='vertical', pos_top='15%', pos_left='2%'))
pie1.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{d}%"))
pie1.set_colors(['#FF7F0E', '#1F77B4', '#2CA02C', '#D62728', '#946C8B'])

#显示图形
pie1.render_notebook()
```

```python
#(11)按照区域绘制条形图
region_name = df_all.groupby('区域')['总价格（万）'].count().index.tolist()
#region_name
house_num = df_all.groupby('区域')['总价格（万）'].count().values
#house_num

# 数据框
df_region = pd.DataFrame({
    'region_name': region_name,
    'house_num': house_num
})
df_region = df_region.sort_values('house_num', ascending=False)
#df_region

# 产生数据
x_data = df_region['region_name'].values.tolist()
y_data = df_region['house_num'].values.tolist()

# 绘制条形图
bar1 = Bar(init_opts=opts.InitOpts(width='900px', height='550px'))
bar1.add_xaxis(x_data)
bar1.add_yaxis('', y_data)
bar1.set_global_opts(title_opts=opts.TitleOpts(title='广州不同区域的二手房房源数量'), 
                     visualmap_opts=opts.VisualMapOpts(max_=1300))

#显示图形
bar1.render_notebook()
```

```python
#（12）广州二手房面积和总价的关系
plt.scatter(x=df_all['面积'], y=df_all['总价格（万）'])
```

### 4）聚类分析

```python
#导包
from sklearn.cluster import KMeans
from sklearn.cluster import MeanShift, estimate_bandwidth
X = df_all.iloc[:, [3,11]].values#行：，[第六列和第七列],x=3，y=11
#print(type(X))
#print(X.shape)
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)#init一种可以避免选择较差起始点方法
y_kmeans=kmeans.fit_predict(X)
print(y_kmeans)
```

```python
kmeans.cluster_centers_#五个中心点，两列
```

```python
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.scatter(X[y_kmeans== 0,0],X[y_kmeans == 0,1],s = 50,c = 'pink',label = 'one')#s为点的大小，c为点的颜色，标签
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1],s = 50,c = 'blue',label = 'tow')
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1],s = 50,c = 'green',label = 'three')
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1],s = 50,c = 'red',label = 'four')
plt.scatter(X[y_kmeans == 4,0],X[y_kmeans == 4,1],s = 50,c = 'yellow',label = 'five')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,color='black',label='centroids')
plt.title('每平方价格与建造时长')
plt.xlabel('单价（元/平方米）')
plt.ylabel('建筑年份')
plt.legend()
plt.show()
```

#### 聚类分析：通过聚类分析可以发现，单价在400元/平方米的建造年份大多处于3-6年，而价格越高则越稳定在一个值。由此可见，每平方价格越高，建筑年份处于稳定区域，因此，选择价格高的房子性价比还是很高的。
